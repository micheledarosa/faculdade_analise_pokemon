## Acesso aos dados 

Lendo o arquivo e formas de mostrar a tabela


```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
display(df)

# display(df.columns)  # mostra o nome das colunas da tabela
# display(df.head(3))  # mostra os primeiros 3 dados da tabela
# display(df.tail(2))  # mostra os últimos 2 dados da tabela
```

Acessando colunas individuais

```python
# df['Nome']
# df[['Nome', 'Tipo 1']]  # acessando mais de uma coluna
df[['Nome', 'Tipo 1']][0:150:2]  # do 1º até o 150º pulando de 2 em 2
```

Acessando linhas individuais

```python
df.iloc[:4]  # do começo até o 3
```

Acessando linhas e colunas individuais

```python
df.iloc[2, 1]  #  coluna 1, posição 2
```

Buscando por uma informação específica

```python
df.loc[df['Tipo 1'] == 'Fogo']
```

Acessando as linhas por nome

```python
for indice, linha in df.iterrows():
    print(indice, linha['Nome'])
    if indice == 150:  # if linha['#'] == 150: 
        break
```
