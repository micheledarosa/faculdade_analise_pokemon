# Ordenação



```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
```

Ordenar por ordem alfabética

```python
# df.sort_values('Nome')
df.sort_values('Nome', ascending=False)
```

Realizando mudanças nos dados

```python
df['Total'] = df['Ataque'] + df['Atq Esp'] + df['Def Esp'] + df['Defensa'] + df['Velocidade'] + df['Vida']
df
```

Removendo uma coluna

```python
df = df.drop(columns=['Total'])
df
```

Forma alternativa de criar coluna Total

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1)  # todas as linhas e da coluna 4(vida) até a coluna 9(velocidade)
df
```

Renomeando a coluna

```python
df.rename({'Defensa':'Defesa'}, axis=1, inplace=True)  # axis = eixo 1 (colunas)
df
```

Reordenando coluna

```python
# df2 = df[['Total', 'Vida', 'Defesa']]
# df2

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df
```

Salvar arquivos

```python
df.to_csv('modificado.csv', index=False)  # para não criar uma coluna de índice

# df.to_csv('modificado.csv')
# df.to_excel('modificado.xlsx', index=False)
# df.to_csv('modificado.txt', index=False, sep='\t')  # separa por tab e não vírgula
```

Filtrando dados

- & = E
- | = OU

```python
novo_df = df.loc[(df['Tipo 1'] == 'Fogo') & (df['Tipo 2'] == 'Agua')]
novo_df.reset_index(drop=True)
```

```python
novo_df = df.loc[(df['Tipo 1'] == 'Fogo') | (df['Tipo 2'] == 'Agua')]
novo_df.reset_index(drop=True, inplace=True)
novo_df
```

Filtrando versões Mega

[~df] = não contém a palavra 'Mega '

```python
novo_df = df.loc[~df['Nome'].str.contains('Mega ')]
novo_df
```

Expressões regulares para pokémons de água e fogo

- regex = True indica que está usando as expressões regulares
- flags=re.I (ignore case) indicando que está ignorando letras maiúsculas e minúsculas

```python
import re

novo_df = df.loc[df['Tipo 1'].str.contains('fogo|agua', flags=re.I, regex=True)]
novo_df
```

Procurando pokémons que começam com "pi"

- ^pi = começa com pi
- pi$ = termina com pi

```python
novo_df = df.loc[df['Nome'].str.contains('^pi', flags=re.I, regex=True)]
novo_df
```

Mudanças condicionais

[df['Tipo 1'] == 'Fogo', 'Tipo 1'] depois da vírgula é a coluna que deseja alterar

```python
df.loc[df['Tipo 1'] == 'Fogo', 'Tipo 1'] = 'Fogueteiro'
df
```

```python
# o pokémon que passar de 500 no total é lendário
df.loc[df['Total'] > 500, 'Lendario'] = True
df
```

```python
df.loc[df['Velocidade'] > 70, ['Lendario', 'Geracao']] = ['Teste1', 'Teste2']
df.head(11)
```

Plotagem de dados

histograma

```python
import matplotlib.pyplot as plt 
import pandas as pd

atributo = 'Ataque'
bins = range(0, 201, 20)
plt.xticks(bins)  # xticks = valores que mostra no eixo x e y

df = pd.read_csv('modificado.csv')

plt.title('Distribuição dos pokémons por ' + atributo)
plt.xlabel('Valor do atributo ' + atributo)
plt.ylabel('Quantidade de pokémons')
plt.hist(df[atributo], bins=bins)
plt.show()
```
